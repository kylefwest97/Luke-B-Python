```python
import adjustText as adjust_text
import pyjokes
import sys

# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 
import seaborn as sns
# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
#Filter Down to US Jobs

df_us_da = df[(df['job_country']=='United States')&(df['job_title_short']=='Data Analyst')].copy()
df_us_da
```

```python
#EDA

loc_count = df_us_da['job_location'].value_counts().head(10).reset_index()

sns.barplot(data=loc_count, y= 'job_location',x='count', hue='count', legend=False, palette='dark:salmon_r')
plt.title("Data Analyst Job Postings by Location")
plt.ylabel('')
plt.xlabel('Job Count')
```

```python
import numpy as np
pie = ['job_health_insurance', 'job_no_degree_mention', 'job_work_from_home']
title = ['Health Insurance', 'Degree Mentioned', 'Work From Home']
labels = ['No', 'Yes']

fig, ax = plt.subplots(1,3, figsize = (10,4))

fig.suptitle('Data Analyst Jobs in the US')



for i, cat in enumerate(pie):
    plot = df_us_da[cat].value_counts().reset_index()
    colors = plt.get_cmap('Set1')(np.linspace(0.2, 0.7, len(plot)))
    ax[i].pie(data=plot, x='count', labels = labels, autopct='%.0f%%', colors = colors)
    ax[i].set_title(title[i])
    
    
   
 
    

```

```python
company_count = df_us_da['company_name'].value_counts().head(10).reset_index()

sns.barplot(data=company_count, y= 'company_name',x='count', hue='count', legend=False, palette='dark:salmon_r')
plt.title("Data Analyst Job Postings by Company")
plt.ylabel('')
plt.xlabel('Job Count')
```
